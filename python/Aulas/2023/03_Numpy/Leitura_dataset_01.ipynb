# Leitura de Dataset

```python
import numpy as np
```

```python
# Array de 1 dimensão
array_pokemon = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=5)
print(array_pokemon)
```

```python
# Array de 2 dimensões
array_pokemon = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=[5,6])
print(array_pokemon)
```

```python
array_nomes = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=1, dtype='str')
print(array_nomes)
```

```python
print(array_nomes.size)
```

```python
for nome in array_nomes:
    print(nome)
```
