# Pandas - Operações

```python
import pandas as pd
```

```python
dataframe_1 = pd.read_csv('Pokemon.csv')
dataframe_1.head(10)
```

### Média

Função mean()

```python
dataframe_1['HP'].mean()
```

```python
dataframe_1.HP.mean()
```

### Mediana

Função median()

```python
dataframe_1.Attack.median()
```

### Desvio Padrão

Função std()

```python
dataframe_1['Defense'].std()
```

### Operações entre colunas

```python
dataframe_1.HP + dataframe_1.Attack
```

```python
dataframe_1.HP - dataframe_1.Attack
```

```python
dataframe_1.HP * dataframe_1.Attack
```

```python
dataframe_1.HP / dataframe_1.Attack
```

### Adicionando condições às consultas

#### Operadores Condicionais
#### > maior
#### >= maior ou igual
#### < menor
#### <= menor ou igual
#### == igual
#### != diferente


```python
# Difícil visualização e entendimento, pois imprime a condição para cada linha
dataframe_1['Name'] == 'Venusaur'
```

```python
dataframe_1[dataframe_1['Name'] == 'Venusaur']
```

```python
dataframe_1[dataframe_1.Name == 'Venusaur']
```

```python
dataframe_1[dataframe_1.HP == 80]
```

```python
dataframe_1.loc[dataframe_1.HP == 80]
```

```python
# Imprime os nomes dos Pokemons que satisfazem a condição de HP == 80
dataframe_1.loc[dataframe_1.HP == 80, ['Name']]
```

```python
dataframe_1.loc[dataframe_1.HP == 80, ['Name','Type 1']]
```

```python
# Todos os pokemons que o valor de HP é maior do que o seu Attack
dataframe_1.loc[dataframe_1.HP > dataframe_1.Attack]
```

```python
# Todos os pokemons que possuem HP maior que a média de HP de todos
dataframe_1.loc[dataframe_1.HP > dataframe_1.HP.mean()]
```

#### Operadores Lógicos
#### & e
#### | ou
#### ~ negação

```python
# Imprimir todos que possuem HP menor que 50 ou maior que 90
dataframe_1.loc[(dataframe_1.HP<50) | (dataframe_1.HP>90)]
```

```python
# Imprimir todos os que possuem o Defense entre 50 e 80
dataframe_1[(dataframe_1.Defense>50) & (dataframe_1.Defense<80)]
```

```python
# Imprimir todos os que possuem HP maior que 60 e Defense menor que 40
dataframe_1[(dataframe_1.HP>60) & (dataframe_1.Defense<40)]
```

#### Condição e definindo quais colunas deve exibir

```python
dataframe_1.loc[(dataframe_1.HP>60) & (dataframe_1.Defense<40), ['Name', 'Type 1']]
```
