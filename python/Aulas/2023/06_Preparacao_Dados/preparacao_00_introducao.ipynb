# PREPARAÇÃO/LIMPEZA DE DADOS

### Fases da Análise de Dados:
1. Problema a ser resolvido;
2. Preparação e Exploração dos dados;
3. Análise Descritiva e Exploratória dos dados;



### O que é feito na preparação/limpeza de dados:
- Identificação de variáveis (colunas);
- Busca por dados nulos/ausentes (missing);
- Busca por dados duplicados;
- Busca por valores atípicos (outliers);
- Inconsistência entre bases (exemplo: venda para clientes inexistentes);
- Ajustar Formato de dados (ex: data, dinheiro, entre outros);

### Se mais de um dataset e dados estiverem relacionados (exemplo: banco de dados):
- Realizar a junção dos datasets;
    - Identificação de chaves primárias.
- Normalização.
