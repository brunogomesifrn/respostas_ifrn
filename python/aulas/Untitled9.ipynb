# Matplotlib

- Matplotlib é uma biblioteca python para visualização de dados e geração de gráficos, dos mais básicos até os mais avançados;
- Tem uma excelente integração com o jupyter notebook, permitindo a geração de gráficos no próprio jupyter, além de recursos como zoom;
- Permite exportar gráficos para formatos de vetor, pdf, png, jpg entre várias outras;
- É possível personalizar o gráfico e a geração pode ser feita através de funções ou mesmo através de prog. Orientada a Objetos.

Site Oficial: https://matplotlib.org/
        
Instalação: pip install matplotlib  

```python
#Importando a biblioteca principal
import matplotlib as mpl
```

```python
#Importando a biblioteca que facilita a geração dos gráficos
import matplotlib.pyplot as plt
```

```python
#Este comando serve para gerar os gráficos no próprio jupyter
%matplotlib inline
```

```python
plt.plot([1,2,3],[10,5,15])
plt.show()
```

```python
x = [1,2,4]
y = [10,4,2]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,4]
y = [10,4,2]
plt.plot(x,y)
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.show()
```

```python
#label - adiciona um rótulo ao gráfico (não aparece, é opcional)
#legend() - exibe o rótulo/legenda do gráfico
x = [1,2,4]
y = [10,4,2]
plt.plot(x,y, label='Legenda do Gráfico')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.show()
```

```python
#grid() - adiciona as grades do gráfico
x = [1,2,4]
y = [10,4,2]
plt.plot(x,y, label='Legenda do Gráfico')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.grid(True)
plt.show()
```

Cores:
r - vermelho
g - verde
b - azul
c - ciano
y - amarelo
m - magenta
k - preto
w - branco

https://matplotlib.org/3.1.0/gallery/color/named_colors.html

```python
#color -  muda a cor do gráfico que está sendo gerado
x = [1,2,4]
y = [10,4,2]
plt.plot(x,y, label='Legenda do Gráfico', color='r')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.grid(True)
plt.show()
```

```python
#Inserindo 2 linhas no mesmo gráfico
x = [1,2,4]
y1 = [10,4,2]
y2 = [5,10,2]
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
#Inserindo legenda ao gráfico com as 2 linhas e mundando a cor
x = [1,2,4]
y1 = [10,4,2]
y2 = [5,10,2]
plt.plot(x, y1, label='Resultado 1', color='g')
plt.plot(x, y2, label='Resultado 2', color='c')
plt.legend()
plt.show()
```

# Gráfico de Barras

```python
x = [1,2,3]
y = [10,4,2]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [10,4,2]
plt.bar(x,y, label='Legenda do Gráfico', color='r')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.grid(True)
plt.show()
```

```python
#color -  muda a cor do gráfico que está sendo gerado
x = [1,2,3]
y1 = [10,4,10]
y2 = [4,6,5]
plt.bar(x,y1, label='Legenda do Gráfico 1', color='r')
plt.bar(x,y2, label='Legenda do Gráfico 2', color='b')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.grid(True)
plt.show()
```

```python
'''
VERIFICAR:
#x2 = [(x + 0.2) for x in valor]
#x3 = [(x - 0.2) for x in valor]
'''
#Valores antigos
x = [1,2,3]

#x2 aumenta em cada posição 0.2, e x3 diminui 0.2 em cada posição
x2 = [1.2, 2.2, 3.2]
x3 = [0.8, 1.8, 2.8]

y1 = [10,4,10]
y2 = [4,2,5]

plt.bar(x2,y1, label='Legenda do Gráfico 1', color='r', width=0.4)
plt.bar(x3,y2, label='Legenda do Gráfico 2', color='b', width=0.4)
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Título do Gráfico')
plt.grid(True)
plt.show()
```

# Gráfico de Pizza

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Excelente']
votos = [0, 2, 1, 8, 5]
plt.pie(votos, labels=opcoes)
plt.show()
```

```python
#Adicionando legendas e cores
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Excelente']
votos = [0, 2, 1, 8, 5]
cores = ['b', 'red', 'y', 'c', 'coral']
plt.pie(votos, labels=opcoes, colors=cores)
plt.legend()
plt.show()
```

```python
#autopct - adiciona a % em cada fatia do gráfico
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Excelente']
votos = [0, 2, 1, 8, 5]
cores = ['b', 'red', 'y', 'c', 'coral']
plt.pie(votos, labels=opcoes, colors=cores, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Excelente']
votos = [0, 2, 1, 8, 5]
cores = ['b', 'red', 'y', 'c', 'coral']
plt.pie(votos, labels=opcoes, colors=cores, autopct='%1.1f%%', explode=(0,0,0,0,0.2))
plt.legend()
plt.show()
```
