# Pandas + CSV

```python
import pandas as pd
```

```python
'''
Por padrão, o separador que a função read_csv considera é a vírgula ','
Se o csv utilizar outro separador, deve usar a propriedade sep
'''
planilha_1 = pd.read_csv('country_wise_latest.csv')
```

```python
planilha_1
```

```python
'''
Outra forma de leitura
Aceita outros formatos de arquivos de tabelas e obrigatoriamente
precisa informar o separador.
'''

planilha_2 = pd.read_table('country_wise_latest.csv', sep=',')
planilha_2
```

```python
#Slicing da linha 2 até a 4
planilha_2[2:5]
```

```python
#Slicing da linha 4 em diante
planilha_2[4:]
```

```python
#Slicing de Colunas (da coluna Confirmed em diante)
planilha_2.loc[:,'Confirmed':]
```

```python
#Slicing de Colunas (da coluna Confirmed até Active)
planilha_2.loc[:,'Confirmed':'Active']
```

```python
#Slicing de Colunas (Apenas as colunas Confirmed e Recovered)
planilha_2.loc[:,['Confirmed','Recovered']]
```

```python
#Slicing utilizando índices de colunas no lugar do nome
#Exemplo abaixo irá imprimir as colunas 1,2,3 e 4
planilha_2.iloc[:,1:5]
```

```python
#Retorna as colunas de índices 1, 5 e 10
planilha_2.iloc[:,[1,5,10]]
```

```python
#head retorna um resumo (apenas as primeiras linhas)
planilha_2.head()
```

```python
#describe - retorna um resumo estatístico de cada coluna
planilha_2.describe()
```

```python
#mean - média por coluna
# para toda a planilha: planilha_2.mean()

#O exemplo a seguir mostra a média apenas das colunas Confirmed e Deaths
planilha_2.loc[:,'Confirmed':'Deaths'].mean()
```

```python
#std - Desvio padrão
planilha_2.loc[:,'Confirmed':'Deaths'].std()
```

```python
#planilha_2       -   Retorna toda a planilha

#planilha_1[2:5]  - Retorna a planilha fatiada por linha (da 2 até a 5)

#planilha_2.loc[:,'Deaths':'Active']   - Retorna a planilha fat. por nome de coluna

#planilha_2.loc[:,['Deaths','Active']]  - Retorna as colunas da lista de nomes

#planilha_2.iloc[:,1:4] - iloc serve para informar a posição da coluna

#planilha_2.iloc[:,[1,5,10]] - Retorna as colunas de acordo com suas posições
```
